## Matplotlib Markers
#### Markers 

```python
#you can use the keyword argument marker to emphasize each point with a specified marker
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='*')
plt.show()
```

## Marker size
### set the marker size

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=10)
plt.show()
```

### Marker COlor
#### you can sue the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the marker
#### set the EDGE color of red

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=10,mec='r')
plt.show()
```

## Marker Face color
#### you can sue keyword argument markerfacecolor or shorter mfc


```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mfc='r')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mec='r',mfc='k')
plt.show()
```

### Create a title for a plot

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calrie Burnage")
plt.show()
```

### Set Font Properties for Title and Label

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}


plt.title("Sports Watch Data",fontdict=font1)
plt.xlabel("Average Pulse",fontdict=font2)
plt.ylabel("Calrie Burnage",fontdict=font2)
plt.plot(x,y)
plt.show()
```

### Position the title

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}


plt.title("Sports Watch Data",loc='left')
plt.xlabel("Average Pulse",fontdict=font2)
plt.ylabel("Calrie Burnage",fontdict=font2)
plt.plot(x,y)
plt.show()
```

### Line Name
