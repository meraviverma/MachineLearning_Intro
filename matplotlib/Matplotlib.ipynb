## Matplotlib Markers
#### Markers 

```python
#you can use the keyword argument marker to emphasize each point with a specified marker
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='*')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,'o-.r')
plt.show()

#0 = marker level
#:=line value
#r=color reference
#'-' solid line
# ':' '--' '-.'

# 'r','g','b','c','m','y','k','w'
```

## Marker size
### set the marker size

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=10)
plt.show()
```

### Marker COlor
#### you can sue the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the marker
#### set the EDGE color of red

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=10,mec='r')
plt.show()
```

## Marker Face color
#### you can sue keyword argument markerfacecolor or shorter mfc


```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mfc='r')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mec='r',mfc='k')
plt.show()
```

## Matplotlib line
### you can use keyword argument linestyle, or shorter is , to change the style of the plotted line

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linestyle='dotted')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linestyle='dashed')
plt.show()
```

## shorter syntax
### linestyle is.
### dotted as :
### dashed as --

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,ls=':')
plt.show()
```

### solid '-'
### dotted ':'
### dashed '--'
### dashdot '-.'
### None "or''

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,c='r')
plt.show()
```

## Line Width

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linewidth='20.5')
plt.show()
```

## Multiple Lines

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1=np.array([0,1,2,3])
x2=np.array([3,8,1,10])
y1=np.array([0,1,2,3])
y2=np.array([6,2,7,11])
plt.plot(x1,x2,y1,y2)
plt.show()
```
