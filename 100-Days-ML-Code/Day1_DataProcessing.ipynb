## Handling Missing Data

### SimpleImputer is imported from scikit-learn to handle missing values.

missing_values=np.nan: tells the imputer that np.nan marks missing entries.

strategy="mean": missing values will be replaced with the mean of the column.

X[:, 1:3] selects columns 1 and 2 (Python indexing is zero-based).

The imputer calculates the mean of each selected column, ignoring np.nan.

The imputer replaces all np.nan values in those columns with the computed column means.

The transformed values are written back into the original dataset X.

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Create the imputer to replace NaN with column mean
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")

# Fit on columns 1 and 2
imputer = imputer.fit(X[:, 1:3])

# Transform and replace
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

## Step 4: Encoding categorical data

Importing encoders

LabelEncoder: Converts categorical labels (like "red", "blue", "green") into numeric codes (0, 1, 2).

OneHotEncoder: Converts categorical labels into binary vectors (e.g., "red" → [1,0,0], "blue" → [0,1,0]).

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X[ : , 0] = labelencoder_X.fit_transform(X[ : , 0])
```

```python
X
```

Creating a dummy variable

```python
#onehotencoder = OneHotEncoder(categorical_features = [0])
# Apply OneHotEncoder only to column 0
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [0])],
    remainder='passthrough'   # keep other columns unchanged
)
X = ct.fit_transform(X).astype(int)

labelencoder_Y = LabelEncoder()
Y =  labelencoder_Y.fit_transform(Y)
```
