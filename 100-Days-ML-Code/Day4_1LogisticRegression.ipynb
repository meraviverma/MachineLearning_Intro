```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
X=np.array([[100],[110],[123],[207],[236]])
Y=np.array([0,0,0,1,1])
```

```python
model=LogisticRegression()
model.fit(X,Y)
```

```python
# Generate a range of values for plotting
x_range = np.linspace(90, 250, 300).reshape(-1, 1)

# Predict probabilities
y_prob = model.predict_proba(x_range)[:, 1]

# Plot data points
plt.scatter(X, Y, color="black", label="Training data")

# Plot logistic regression curve
plt.plot(x_range, y_prob, color="blue", linewidth=2, label="Logistic curve")

# Plot decision boundary (p=0.5 cutoff)
decision_boundary = x_range[np.argmin(np.abs(y_prob - 0.5))][0]
plt.axvline(decision_boundary, color="red", linestyle="--", label=f"Decision boundary ≈ {decision_boundary:.2f}")

# Labels and legend
plt.xlabel("X values")
plt.ylabel("Probability of class 1")
plt.title("Logistic Regression on 1D Data")
plt.legend()
plt.show()
```
